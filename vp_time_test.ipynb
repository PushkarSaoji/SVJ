```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from math import*
```

```python
col_names= ["N", "tree_t", "search_t"]

data= pd.read_csv("SIFT/tree_time.csv", names= col_names)
```

```python
x= data["N"]
y1=data["tree_t"]
y2=data["search_t"]
```

```python
data
```

```python
a1, b1 = np.polyfit(x, y1, 1)
y_fit1= a1*x + b1
```

```python
a2, b2 = np.polyfit(x, y2, 1)
y_fit2= a2*x + b2
```

```python
plt.scatter(x, y1)
plt.plot(x, y_fit1, label=f"Fit: y =  {a1:.2f}x + {b1:.2f}", color="blue")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("poly Fit")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
plt.scatter(x, y2)
plt.plot(x, y_fit2, label=f"Fit: y =  {a2:.2f}x + {b2:.2f}", color="blue")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("poly Fit")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
